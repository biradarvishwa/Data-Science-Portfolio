```python
# State Hypotheses
Null="The average weekly cost equals the model prediction."
Alternative="The average weekly cost is higher than the model prediction."
print("Null_Hypotheses(H0):",Null)
print("Alternative_Hypothese(H1):",Alternative)
```

```python
import pandas as pd
import numpy as np
# t-tests
import scipy.stats as stats
from scipy.stats import norm
from scipy.stats import ttest_1samp, ttest_ind, ttest_rel

# z-tests
import statsmodels.api as sm
```

```python
sample_mean=3050 
X_mean=600
sigma_units=25
sigma_cost=5
n=25
sample_size=25
sigma=sigma_cost*sigma_units

#Calculate theoretical mean
pop_mean=1000+5*X_mean        #->W = $1,000 + $5X

# calculate Z_score
# Z = (x̄ - μ) / (σ / √n)

z_score = (sample_mean - pop_mean) / (sigma / np.sqrt(n))
print("Z-score:", z_score)
```

```python
alpha = 0.05
z_score=-38.0

# calculate critical value 
critical_value=norm.ppf(1 - alpha)
print('critical_value:',critical_value)

# calculate p-value
p_value= 1-stats.norm.cdf(z_score)
print('P-value:', p_value)

# compare p-value with alpha
if p_value<alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to reject the null hypothesis:","their is no significant evidence for the observed weekly operating costs are higher.")
```
