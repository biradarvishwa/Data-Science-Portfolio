Exercise 1 
Prime Numbers

```python
# Program to check if a number is prime

num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    is_prime = True
    
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

 Exercise 2 Product of Random numbers

```python
import random

#Generating Random number
num1=random.randint(1,10)
num2=random.randint(1,10)

#Taking the answer from the user 
print(f'What is the product of {num1} and {num2}?')
answer = int(input("Enter the Product of two random numbers"))

#Checking the answer given by user is correct or not
correct_answer = num1 * num2
if answer == correct_answer:
    print("The answer is correct")
else:
    print(f"The Answer is Incorrect, The correct answer is:{correct_answer}")
```

Squares of Even/Odd numbers

```python
# Program to print squares of EVEN numbers from 100 to 200

# Choice: EVEN numbers
# Reason: Even numbers are divisible by 2 (num % 2 == 0)

for num in range(100, 201):  # 200 included
    if num % 2 == 0:         # Check if number is even
        square = num ** 2
        print(f"{num}^2 = {square}")
```

# Exercise 4
# Hospital Billing System 

```python
#Given Values
room_charges_per_day = 2000
doc_fee = 1500
lab_test_charge = 300

# Patient details
days_stayed = 4
lab_tests = 3
medicine_charges = 2400

room_charges = days_stayed * room_charges_per_day
lab_charges = lab_tests * lab_test_charge

#Total bill before discount
total_bill = room_charges + lab_charges + medicine_charges + doc_fee

if total_bill>10000:
    discount = 0.10 * total_bill
else:
    discount = 0

print("----- Hospital Bill -----")
print(f"Room Charges: ₹{room_charges}")
print(f"Doctor Fee: ₹{doc_fee}")
print(f"Lab Charges: ₹{lab_charges}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Total (before discount): ₹{total_bill}")
print(f"Discount: ₹{discount}")
print(f"Final Bill: ₹{total_bill}")
```
