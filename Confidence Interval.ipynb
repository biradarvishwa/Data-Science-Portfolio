```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
n=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07,1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
sample_size=len(n)
mean=np.mean(n)
std=np.std(n)

df=sample_size-1

t_critical=stats.t.ppf(0.995,df)

margin_error=t_critical*(std/np.sqrt(sample_size))


lower=mean-margin_error
upper=mean+margin_error

CI_sample=(lower,upper)

print("99% Confidence Interval (t-distribution):",CI_sample)
```

```python
n=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07,1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
sample_size=len(n)
std=0.2

z_critical=stats.norm.ppf(0.995)
margin_error=z_critical*(std/np.sqrt(sample_size))


lower=mean-margin_error
upper=mean+margin_error

CI_known=(lower,upper)

print("99% Confidence Interval (z-distribution):",CI_known)
```
