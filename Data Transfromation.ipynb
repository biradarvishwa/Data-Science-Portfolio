```python
# Replace ? with NaN
df.replace(' ?', np.nan, inplace=True)
df.replace('?', np.nan, inplace=True)

# Check missing values
print(df.isnull().sum())
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
num_cols = ['age',
            'fnlwgt',
            'education_num',
            'capital_gain',
            'capital_loss',
            'hours_per_week']
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
standard_scaler = StandardScaler()

df_standard = df.copy()

df_standard[num_cols] = standard_scaler.fit_transform(
    df_standard[num_cols]
)

df_standard[num_cols].head()
```

```python
minmax_scaler = MinMaxScaler()

df_minmax = df.copy()

df_minmax[num_cols] = minmax_scaler.fit_transform(
    df_minmax[num_cols]
)

df_minmax[num_cols].head()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(col, ":", df[col].nunique())
```

```python
low_cardinality = []

for col in cat_cols:
    if df[col].nunique() < 5:
        low_cardinality.append(col)

print(low_cardinality)
```

```python
df_ohe = pd.get_dummies(
    df,
    columns=low_cardinality,
    drop_first=True
)

df_ohe.head()
```

```python
high_cardinality = []

for col in cat_cols:
    if df[col].nunique() >= 5:
        high_cardinality.append(col)

print(high_cardinality)
```

```python
df_label = df.copy()

le = LabelEncoder()

for col in high_cardinality:
    df_label[col] = le.fit_transform(df_label[col])

df_label.head()
```

```python
df['capital_difference'] = (
    df['capital_gain'] - df['capital_loss']
)
```

```python
df['age_hours_product'] = (
    df['age'] * df['hours_per_week']
)
```

```python
df[['capital_difference',
    'age_hours_product']].head()
```

```python
print(df[num_cols].skew())
```

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

```python
print("Before Log:",
      df['capital_gain'].skew())

print("After Log:",
      df['capital_gain_log'].skew()) 

#Log Transformation
#Applied to capital_gain because it is highly right-skewed. Log transformation reduces skewness and improves model performance.
```
