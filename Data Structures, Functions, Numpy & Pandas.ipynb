### Section A: Data Structures & Control Structures ###

```python
a=[5,4,3,2,1]
print("Orginal List",a)

a.sort()
print("Sorted List",a)

a.append(6)
print("Appended List",a)

a.extend([7,8])
print("Extended List",a)

a.insert(8,9)
print("Inserted List",a)

a.remove(2)
print("Removed List",a)

a.pop()
print("Pop List",a)

##INDEX##
print("Index",a.index(4))

##COUNT##
print("count",a.count(8))

##REVERSE LIST
print("Reveresed List",a[::-1])

##CLEARED LIST##
a.clear()
print("clearded List",a)
```

```python
students=("Neela","Divya","Sangu")
print(type(students))
print("Orginal Tuple",students)

print(students[1])

students[2]="Sahana"
'''--> Tuples are immutable data types in python it means we cannot do the following.
a) ADD
b) DELETE
c) MODIFY 
--> It is commonly called as read only version of LIST , Indexing is possible'''
```

```python
a={21,24,1,12,5,28,21,1,5}
print(type(a))
print("SET",a)

# a) SET is Unordered
# b) SET stores UNIQUE elements only
# c) SET automatically deletes duplicates
```

```python
def number(n):
    if n>0:
        if n%2==0:
            print("The num is positive and Even",n)
        else:
            print("The num is positive and odd",n)
    elif n==0:
        print("The num is Zer",n)
    else:
        print("The num is negative",n)
number(-2)
```
